## Price

```python
df['price'][:5]
```

```python
df['price'] = df['price'].apply(lambda x: x[1:])
```

```python
df['price'] = df['price'].astype(float)
```

## Stock

```python
df['stock'] = df['stock'].apply(lambda x: 'Yes' if x == 'In stock' else 'No')
```

```python
df['stock'].unique()
```

## Availability

```python
df['availability'] =  df['availability'].apply(lambda x: re.findall(r"\d+", x)[0])
```

```python
df['availability']
```

```python
df['availability'] = df['availability'].astype(int)
```

```python
df.head(3)
```

## title

```python
def clean_titles(value):
    cleaned_value = re.sub(f"â\x80\x99s", "", value)
    cleaned_from_parentheses = re.sub(r"\s*#\d+(-\d+)?\s*", "", cleaned_value)
    return cleaned_from_parentheses
```

```python
df['title'] = df['title'].apply(clean_titles)
```

```python
df['title'].unique()
```

## description

```python
df['description'].unique()[:3]
```

```python
def clean_description_feature(value):
    if isinstance(value, str):
        individual_sentences = value.split('.')
        for sentence in individual_sentences:
            sentence.capitalize()
        capitalize_value = '.'.join(individual_sentences)

        clean_text = re.sub(r"â\x80\x99[a-zA-Z]?", "", capitalize_value)
        clean_text = re.sub(r"Â\xa0 â", "", capitalize_value)
        clean_text = re.sub(r"Ã¢", "", clean_text)
        cleaned_text = re.sub(r"[^\w\s,.]", "", clean_text)
        cleaned_more = re.sub(r"...more", "", cleaned_text)
        return cleaned_more
    else:
        return np.nan
```

```python
df['description'] = df['description'].apply(clean_description_feature)
```

```python
df.head()
```

```python
df = df[['title', 'genre', 'description', 'stock', 'availability', 'price']]
```
